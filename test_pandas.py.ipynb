```python
pop_seoul = pd.read_csv("pop_seoul.txt", sep='\t', header = 2)
pop_seoul.head()
```

```python
pop_seoul = pop_seoul.loc[:, ['자치구', '계', '계.1', '계.2', '65세이상고령자']]
pop_seoul.rename(columns={pop_seoul.columns[0] : '구별',
                          pop_seoul.columns[1] : '인구수',
                          pop_seoul.columns[2] : '한국인',
                          pop_seoul.columns[3] : '외국인',
                          pop_seoul.columns[4] : '고령자'}, inplace=True)
pop_seoul.head()
```

```python
pop_seoul.drop([0], inplace=True)
```

```python
for col in pop_seoul.columns[1:] :
    print(col)
    if (isinstance(pop_seoul[col][1], str)):
        print(col)
        pop_seoul[col] = pop_seoul[col].str.replace(',', '').astype('int64')
pop_seoul.head()
```

```python
pop_seoul['외국인비율'] = pop_seoul['외국인']/pop_seoul['인구수']*100
pop_seoul['고령자비율'] = pop_seoul['고령자']/pop_seoul['인구수']*100
pop_seoul.head()
```
