```python
pop_seoul = pd.read_csv("pop_seoul.txt", sep='\t', header = 2)
pop_seoul.head()
```

```python
pop_seoul.drop([0], inplace=True)
```

```python
for col in pop_seoul.columns[1:] :
    print(col)
    if (isinstance(pop_seoul[col][1], str)):
        print(col)
        pop_seoul[col] = pop_seoul[col].str.replace(',', '').astype('int64')
pop_seoul.head()
```

```python
pop_seoul['외국인비율'] = pop_seoul['외국인']/pop_seoul['인구수']*100
pop_seoul['고령자비율'] = pop_seoul['고령자']/pop_seoul['인구수']*100
pop_seoul.head()
```

```python
fp1 = np.polyfit(dr['인구수'], dr['총계'], 1)
f1 = np.poly1d(fp1)
fx = np.linspace(100000, 700000, 100)

dr['오차'] = np.abs(dr['총계'] - f1(dr['인구수']))
dr_sort = dr.sort_values(by='오차', ascending = False)
dr_sort.head()
```

```python
plt.figure(figsize=(14,10))
plt.scatter(dr['인구수'], dr['총계'], c = dr['오차'], s = 50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
for n in range(10) : 
    plt.text(dr_sort['인구수'][n]*1.02, dr_sort['총계'][n]*0.98,
            dr_sort.index[n], fontsize=15)
plt.xlabel('People')
plt.ylabel('CCTV')
plt.colorbar()
plt.grid()
plt.show()
```
